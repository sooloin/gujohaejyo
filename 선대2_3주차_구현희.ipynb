```python
#점 그리기
import numpy as np
import matplotlib.pyplot as plt

a=np.array([4,3])
#Figure 객체를 만든 후
fig=plt.figure()
#1x1개의 서브플롯을 추가, 첫 번째 서브플롯을 사용
ax=fig.add_subplot(1,1,1)

#변수 a의 값을 좌표로 사용하여 점을 그린다.
#a[0]이 x좌표, a[1]이 y좌표, s는 점의 크기
ax.scatter(a[0],a[1],s=30)
#점 옆에 문자 a를 표시, size는 문자의 크기
ax.text(a[0]+0.2, a[1]+0.2, 'a',size=15)
#x축, y축 눈금 범위를 -5~5로 지정
ax.set_xticks(range(-5,6))
ax.set_yticks(range(-5,6))
#그리드를 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)
#x축과 y축 눈금의 비율을 1:1로 함
#참의 크기를 변경해도 비율이 유지됨
ax.set_aspect('equal',adjustable='box')
#다음 코드를 사용하여 원점에 좌표축이 보이게함
#왼쪽 y축과 아래 x축을 (0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')
plt.show()
```

```python
#화살표 그리기
import numpy as np
import matplotlib.pyplot as plt

a=np.array([4,3])
#Figure 객체를 만든 후
fig=plt.figure()
#1x1개의 서브플롯을 추가, 첫 번째 서브플롯을 사용
ax=fig.add_subplot(1,1,1)

#a[0]과 a[1]이 각각 끝 위치의 x좌표, y좌표
ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy', scale=1)
ax.text(a[0], a[1], 'a', size=15)
#x축, y축 눈금 범위를 -5~5로 지정
ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
#그리드를 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)
#x축과 y축 눈금의 비율을 1:1로 함
#창의 크기를 변경해도 비율이 유지됨
ax.set_aspect('equal',adjustable='box')
#다음 코드를 사용하여 원점에 좌표축이 보이게함
#왼쪽 y축과 아래 x축을(0,0)으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_position('zero')
ax.spines['top'].set_position('zero')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터 A와 B를 생성합니다.
A = np.array([1, -2])
B = np.array([4, 3])

# 1개의 행과 2개의 열을 가지는 서브플롯을 생성
f, ax = plt.subplots(1, 2)
ax[0].title.set_text('A')
ax[1].title.set_text('B')

# 각 서브플롯에 시작점이 (0,0)인 벡터를 그린다.
ax[0].quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1)

# 서브플롯별로 그리드를 생성
start_x = -1
end_x = 5
start_y = -3
end_y = 4
for i in range(2):
    ax[i].axis([start_x, end_x, start_y, end_y])
    ax[i].set_xticks(range(start_x, end_x))
    ax[i].set_yticks(range(start_y, end_y))
    ax[i].grid(True)
    ax[i].set_aspect('equal', adjustable='box')

plt.show()
```

```python
#벡터의 성분
import numpy as np
import matplotlib.pyplot as plt

i=np.array([1,0])
j=np.array([0,1])

#단위 벡터로 벡터[4,3]을 나타낼 수 있다.
a=4*i+3*j

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy',scale=1)
ax.text(a[0],a[1],'4i+3j',size=15)

ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show
```

```python
#벡터의 기본 연산 (덧셈) 10p
import numpy as np
import matplotlib.pyplot as plt

a=np.array([4,1])
b=np.array([-2, 3])
sum=a+b

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
#vector a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(a[0], a[1],'a', size=15)

#vector b를 vector a의 끝점에서 시작하도록 그린다.
ax.quiver(a[0], a[1],b[0],b[1],angles='xy', scale_units='xy', scale=1, color='green')
ax.text(a[0]+b[0], a[1]+b[1], 'b', size=15)

#vector b를 vector a의 끝점에서 시작하도록 그린다.
#red는 결과이다.
ax.quiver(0,0,sum[0],sum[1],angles='xy', scale_units='xy', scale=1, color='red')
ax.text(sum[0]*0.5+0.3, b[1], 'a+b', size=15, color='blue')

ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
#벡터의 기본 연산 (뺄셈)
import numpy as np
import matplotlib.pyplot as plt

a=np.array([4,1])
b=np.array([-2,3])
sum=a-b

fig=plt.figure()
ax=fig.add_subplot(1,1,1)
#vector a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(a[0], a[1],'a', size=15)

#vector b를 vector a의 끝점에서 시작하도록 그린다.
ax.quiver(a[0], a[1],-b[0],-b[1],angles='xy', scale_units='xy', scale=1, color='green')
ax.text(a[0]-b[0], a[1]-b[1], 'b', size=15)

#vector b를 vector a의 끝점에서 시작하도록 그린다.
#red는 결과이다.
ax.quiver(0,0,sum[0],sum[1],angles='xy', scale_units='xy', scale=1, color='red')
ax.text(sum[0]*0.5+0.3, b[1], 'a-b', size=15, color='blue')

ax.set_xticks(range(-2,8))
ax.set_yticks(range(-4,5))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
#벡터의 기본 연산 (스칼라곱)
import numpy as np
import matplotlib.pyplot as plt

a=np.array([4,1])
c=3
ca=np.multiply(c,a)

fig=plt.figure()
ax=fig.add_subplot(1,1,1)

#vector a를 원점에서 시작하도록 그린다.
ax.quiver(0,0,a[0],a[1],angles='xy', scale_units='xy', scale=1, color='blue')
ax.text(a[0], a[1],'a', size=15)

#벡터a에 3을 곱한 결과를 그린다. 시작점을(0,1)로 이동하여 그렸다.
ax.quiver(0,1,ca[0],ca[1], angles='xy', scale_units='xy', scale=1, color='green')
ax.text(ca[0],ca[1]+1,'3a',size=15, color='blue')

ax.set_xticks(range(-2,15))
ax.set_yticks(range(-3,9))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
#내적(dot product)
#내적의 계산 결과는 스칼라이다.
#두 벡터에 대해 내적 계산을 위해 'dot'함수를 제공
import numpy as np
a=np.array([1,2])
b=np.array([2,3])
print(np.dot(a,b))

#벡터의 크기를 계산하는 norm함수
a=np.array([1,2])
print(np.linalg.norm(a))
print(np.sqrt(5))
```

```python
#내적(dot product)2

#넘파이를 사용하여 두 값이 같은지 확인
import numpy as np
a=np.array([1,2])
#자기 자신에 대한 내적의 제곱근
print(np.sqrt(np.dot(a,a)))
#벡터의 크기
print(np.linalg.norm(a))

#내적을 구할 수 있는 2가지 방법
a=np.array([3,3])
b=np.array([3,0])
#첫 번째 방법
print(np.dot(a,b))
#두 번째 방법
norm_a=np.linalg.norm(a)
norm_b=np.linalg.norm(b)
print(norm_a*norm_b*np.cos(45*np.pi/180))
```

```python
#외적

#cross함수를 이용하여 외적을 계산.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
a=np.array([5,3,0])
b=np.array([1,2,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#벡터 a를 그린다.
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1], a[2], 'a', size=15)

#벡터b를 그린다.
ax.quiver(0,0,0,b[0],b[1],b[2],color='red',arrow_length_ratio=0.1)
ax.text(b[0],b[1], b[2], 'b', size=15)
#벡터a와 b의 외적을 그린다.
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue',arrow_length_ratio=0.1)
ax.text(c[0],c[1], c[2], 'c', size=15)
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_xlabel('Y')
ax.set_xlabel('Z')
ax.view_init(elev=20,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
plt.show()
```

```python
#외적2
#두 벡터가 평행하면 외적은 0이다.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

a=np.array([3,3,0])
b=np.array([6,6,0])
#외적 계산
c=np.cross(a,b)
fig=plt.figure()
ax=fig.add_subplot(1,1,1,projection='3d')
#벡터 a를 그린다.
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0],a[1], a[2], 'a', size=15)

#벡터b를 그린다.
ax.quiver(0,0,0,b[0],b[1],b[2],color='red',arrow_length_ratio=0.1)
ax.text(b[0],b[1], b[2], 'b', size=15)
#벡터a와 b의 외적을 그린다.
ax.quiver(0,0,0,c[0],c[1],c[2],color='blue',arrow_length_ratio=0.1)
ax.text(c[0],c[1], c[2], 'c', size=15)
ax.set_xlim(0,7)
ax.set_ylim(0,7)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_xlabel('Y')
ax.set_xlabel('Z')
ax.view_init(elev=20,azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
plt.show()
```

```python
#행렬의 정의(1-1) 17p
import numpy as np
A=np.array([[1,2,3,4], [5,6,7,8]])
#행렬A를 구성하는 행 벡터 출력
print(A[0:])
print(A[0,:])
print(A[1])
print(A[1,:])
#행렬A를 구성하는 열 벡터 출력
print(A[:,0])
print(A[:,1])
print(A[:,2])
print(A[:,3])
```

```python
#행렬의 정의(1-2) _ 행렬의 덧셈/뺄셈 17p
import numpy as np
A=np.array([[1,2], [3,4]])
B=np.array([[4,1], [5,2]])
#행렬A를 구성하는 행 벡터 출력
print(A+B)
print(np.add(A,B))
print(A-B)
print(np.subtract(A,B))
```

```python
#벡터의 전치 19p
import numpy as np
X=np.array([[1,2], [3,4], [5,6]])
print(X)
print(X.T)
print(np.transpose(X))
```

```python
#항등행렬
#넘파이에서는 'eye함수'를 사용하여 항등 행렬을 만들 수 있다.
import numpy as np
I=np.eye(3)
A=np.arange(9).reshape(3,3)
print(A)
#AI=IA=A임을 확인
print((np.dot(A,I)==A).all())
print((np.dot(I,A)==A).all())
```

```python
#역행렬
import numpy as np
A=np.array([[1,2],[3,4]])
invA=np.linalg.inv(A)
print(invA)
#(A역행렬)A=A(A역행렬)=I임을 확인
#파이썬에서 결과값이 근사값으로 출력된다.
#np.round함수를 사용
print(np.dot(A,invA))
print(np.round(np.dot(A,invA)))
print(np.round(np.dot(invA,A)))
```
